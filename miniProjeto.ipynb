Após carregar e diagnosticar os dados, foi possivel ver que os tipos de dados estão corretos, porem ha alguns valores nulos que irei tratar

```python
#tratamento songs
import ast

df_songs["album_name"] = df_songs["album_name"].fillna("Unknown")
df_songs["name"] = df_songs["name"].fillna("Unknown")
df_songs["niche_genres"] = df_songs["niche_genres"].str.replace("[]","No niche genres")
```

```python
#tratamento artists
df_artists["name"] = df_artists["name"].fillna("Unknown")
df_artists[df_artists["id"] == "2ZO5sFPyILDQkFdS4HJPFS"]
```

```python
#Função usada para passar artistas e id dos artistas para lista, para explodir e usar merge futuramente
 
import ast
import pandas as pd

def parse_artists(valor):
    if pd.isna(valor):
        return []

    valor = valor.strip()

    if valor.lower() == "none" or valor == "":
        return []

    # string que representa lista
    if valor.startswith("[") and valor.endswith("]"):
        try:
            return ast.literal_eval(valor)
        except:
            return []

    # string simples → vira lista com 1 artista
    return [valor]

df_songs["artists"] = df_songs["artists"].apply(parse_artists)
df_songs["artists"].apply(type).value_counts()
df_songs["artist_ids"] = df_songs["artist_ids"].apply(parse_artists)
df_songs["artist_ids"].apply(type).value_counts()
df_songs
```

```python
df_analise = df_songs.explode(["artists", "artist_ids"])
df_analise
```

```python
df_artistas_musicas = df_analise.merge(df_artists, left_on="artist_ids", right_on="id", how="left")
#Evitar duplicações de tabelas, poderia ser solucionado no merge antes
df_artistas_musicas = df_artistas_musicas.drop(columns=["artists"]) \
                     .rename(columns={"name": "artist"})
df_artistas_musicas.columns
```

```python
#Criação de tabelas de analise (100 artistas com mais musicas e médias de criterios como dancabilidade, energia e duracao dessas musicas)
df_analise_artistas = df_artistas_musicas.groupby(by="name_y", as_index=False).agg(
    total_music=("id_x", "count"),danceability_media=("danceability", "mean"), energy_media=("energy", "mean"), duration_media=("duration_ms","mean"), popularity_media=("popularity_x", "mean")
).sort_values(by="total_music", ascending=False)
df_analise_artistas[df_analise_artistas["total_music"] > 100]
```

```python
df_artistas_musicas.groupby(by="genre") ["genre"].count().sort_values(ascending=False)
```

```python
pd.set_option('display.float_format', lambda x: f'{x:,.2f}')
df_analise_musicas = df_artistas_musicas.groupby(by="genre", as_index=False).agg(
    total_music=("genre", "count"),
    total_artists_followers_media=("total_artist_followers", "mean"),
    total_artists_popularity_media=("avg_artist_popularity", "mean"),
    popularity_music_media=("popularity_x", "mean")
).sort_values(by="total_music", ascending=False)
df_analise_musicas
```
